```python
class dog:

    def __init__(self,name ,age):
        self.name = name
        self.age = age

class pet(dog):

    def __init__(self, name, age):
        super().__init__(name, age)

def main():
    the_p = pet('tiger',2)
    print(the_p.name)
    print(the_p.age)


if __name__ == "__main__":
    main()
```

```python
class Person(object):
    def __init__(self,name , id_number):
        self.name = name
        self.id_number = id_number

    def displayto(self):
        print(self.name)
        print(self.id_number)

    def detail(self):
        print("my name is {}".format(self.name))
        print("my id number is {}".format(self.id_number))
    

class another(Person):

    def __init__(self,name, id_number,salary , post):
        self.salary = salary
        self.post = post
        super().__init__(name,id_number)
    def detail(self):
        print("my nam is {}".format(self.name))
        print("my salary is {}".format(self.salary))
        print("my id number is {}".format(self.id_number))
        print("my post is {}".format(self.post))

    
```

```python
a = another('manjunath','250000','36945','seniour software engineer',)
```

```python
a.detail()
a.displayto()
```

```python
class bird:
    def intro(self):
        print("they are many forms ")

    def flight(self):
        print("most view the company is fly")


class ostrich(bird):
    def flight(self):
        print("ostrich are flaying is succuess")

class sparrow(bird):
    def flight(self):
        print("sparrow is flying on sky")
    
```

```python
b = bird()
s = sparrow()
o = ostrich()
print(b.flight())
print(s.flight())
print(o.flight)
print(b.intro)
print(s.intro())
```

```python
class base:
    def __init__(self):
        self.name = 'malingray prasanna'
        self.age = None

class derived(base):
    def __init__(self):
        super().__init__()
        print(self.name)
        print(self.age)
```

```python
dd = derived()
print(dd.name)
print(dd.age)
```

```python
class Test:

    def __init__(self, a ,b):
        self.a = a 
        self.b = b
    def __repr__(self):
        print("test a:%s and b:b%s"%(self.a,self.b))
```

```python
class person(object):
    
    def __init__(self,name):
        return self.name

    def isempty(self):
        return False

class employee(person):
    def __init__(self):
        person.__init__(self)
    def isempty(self):
        return True


```

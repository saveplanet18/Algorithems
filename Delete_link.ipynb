```python
class Create_Node:

    def __init__(self,ele):

        self.ele = ele
        self.next = None

class Linked_list:
    def __init__(self):

        self.head = None
    
    def Insertion_at_Begin(self,item):
        new_node = Create_Node(item)
        new_node.next = self.head
        self.head = new_node
    
    def Insertion_at_specific(self,item,pos):
        new_node = Create_Node(item)
        if(self.head == None):
            self.head = new_node
            return
        else:
            temp = self.head
            for i in range(pos-1):
                if(temp.next != None):
                    temp = temp.next
            new_node.next = temp.next
            temp.next = new_node

    def Insertion_at_End(self,item):
        new_node = Create_Node(item)
        if(self.head == None):
            self.head = new_node
            return
        else:
            temp = self.head
            while(temp.next != None):
                temp = temp.next
            temp.next = new_node
            
    def delete_first(self):
        temp = self.head
        if(self.head != None):
            self.head = self.head.next
            temp = None

    def print_show(self):
        temp = self.head
        while(temp != None):
            print(temp.ele)
            temp = temp.next

```
